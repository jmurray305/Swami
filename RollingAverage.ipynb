```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
df = pd.read_csv('./data/MASTER2010_2020.csv',index_col='Unnamed: 0')
```

```python

df['HomeOrAway'] = df['HomeOrAway'].replace(np.nan, 'Home', regex=True)

# subsetDataFrame = df[df['Team_Name'] == 'Indianapolis Colts']
# subsetDataFrame
```

```python
def teamTotalsStats(s):
    if s['HomeOrAway'] == 'Home':
         return s['Home_TotYd']
    else:
        return s['Away_TotYd']
    
def teamQBcomp(s):
    if s['HomeOrAway'] == 'Home':
         return s['Home_QB_Passing_Cmp']
    else:
        return s['Away_QB_Passing_Cmp']
    
def teamQBAtt(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_QB_Passing_Att']
    else:
        return s['Away_QB_Passing_Att']
    
def teamQBPass_Yds(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_QB_Passing_Yds']
    else:
        return s['Away_QB_Passing_Yds']   

def teamQBPass_TD(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_QB_Passing_TD']
    else:
        return s['Away_QB_Passing_TD']   

def teamQBPass_INT(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_QB_Passing_Int']
    else:
        return s['Away_QB_Passing_Int']

def teamQBPass_TD(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_QB_Passing_TD']
    else:
        return s['Away_QB_Passing_TD'] 

def teamQBPass_Long(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_QB_Passing_Lng']
    else:
        return s['Away_QB_Passing_Lng'] 
    
def teamQBPass_Rate(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_QB_Passing_Rate']
    else:
        return s['Away_QB_Passing_Rate'] 
    
def teamRB1_Rush_Yds(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_RB1_Rushing_Yds']
    else:
        return s['A_RB1_Rushing_Yds'] 
      
def teamRB1_Lng(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_RB1_Rushing_Lng']
    else:
        return s['A_RB1_Rushing_Lng'] 
    
def teamRB1_TD(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_RB1_Receiving_TD']
    else:
        return s['A_RB1_Receiving_TD'] 
    
def teamWR1_Tgt(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR1_Receiving_Tgt']
    else:
        return s['A_WR1_Receiving_Tgt'] 
    
def teamWR1_Rec(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR1_Receiving_Rec']
    else:
        return s['H_WR1_Receiving_Rec'] 
    
def teamWR1_Yrds(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR1_Receiving_Yds']
    else:
        return s['A_WR1_Receiving_Yds'] 
    
def teamWR1_TDs(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR1_Rushing_TD']
    else:
        return s['A_WR1_Rushing_TD'] 
    
def teamWR2_Tgt(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR2_Receiving_Tgt']
    else:
        return s['A_WR2_Receiving_Tgt'] 
    
def teamWR2_Rec(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR2_Receiving_Rec']
    else:
        return s['H_WR2_Receiving_Rec'] 
    
def teamWR2_Yrds(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR2_Receiving_Yds']
    else:
        return s['A_WR2_Receiving_Yds'] 
    
def teamWR2_TDs(s):
    if s['HomeOrAway'] == 'Home':
        return s['H_WR2_Rushing_TD']
    else:
        return s['A_WR2_Rushing_TD'] 

def team1stDown(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_First_Downs']
    else:
        return s['Away_First_Downs'] 
    
def team3rdDown(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_Third_Down_Conv']
    else:
        return s['Away_Third_Down_Conv'] 
    
def teamTimePoss(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_Time_of_Possession']
    else:
        return s['Away_Time_of_Possession'] 

def teamPenalties(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_Penalties']
    else:
        return s['Away_Penalties'] 

def teamPenaltiesYrd(s):
    if s['HomeOrAway'] == 'Home':
        return s['Home_Penalties_Yd']
    else:
        return s['Away_Penalties_Yd'] 

```

```python
teamnames = df['Team_Name'].value_counts().keys()
```

```python
ROLLING_MEAN = pd.DataFrame()
for x in range(2018,2021):
    yeardf = pd.DataFrame()
    for i in teamnames:
        try:
            print(i,x)
            rolltemp = pd.DataFrame()
            roll = pd.DataFrame()
            subyear = df[df['Team_Year'] == x ]
            subteam = subyear[subyear['Team_Name'] == i]
            subteam = subteam.sort_values(by=['Date_y'])
            subteam = subteam[subteam['Week'].notna()]
            subteam.drop_duplicates(inplace=True)
            subteam = subteam.reset_index(drop=True)
            roll['teamTotalYrds'] = subteam.apply(lambda row: teamTotalsStats(row), axis=1)
            rolltemp['temp'] = roll['teamTotalYrds']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['TeamTotalYardsRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBcomp'] = subteam.apply(lambda row: teamQBcomp(row), axis = 1)
            rolltemp['temp'] = roll['teamQBcomp']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamQBcompRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBAtt'] = subteam.apply(lambda row: teamQBAtt(row), axis = 1)
            rolltemp['temp'] = roll['teamQBAtt']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['eamQBAttRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBPass_Yds'] = subteam.apply(lambda row: teamQBPass_Yds(row), axis = 1)
            rolltemp['temp'] = roll['teamQBPass_Yds']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamQBPass_YdsRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBPass_TD'] = subteam.apply(lambda row: teamQBPass_TD(row), axis = 1)
            rolltemp['temp'] = roll['teamQBPass_TD']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamQBPass_TDRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBPass_INT'] = subteam.apply(lambda row: teamQBPass_INT(row), axis = 1)
            rolltemp['temp'] = roll['teamQBPass_INT']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamQBPass_INTRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBPass_TD'] = subteam.apply(lambda row: teamQBPass_TD(row), axis = 1)
            rolltemp['temp'] = roll['teamQBPass_TD']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamQBPass_TDRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBPass_Long'] = subteam.apply(lambda row: teamQBPass_Long(row), axis = 1)
            rolltemp['temp'] = roll['teamQBPass_Long']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamQBPass_LongRoll'] = rolltemp.iloc[:,-1]
            roll['teamQBPass_Rate'] = subteam.apply(lambda row: teamQBPass_Rate(row), axis = 1)
            rolltemp['temp'] = roll['teamQBPass_Rate']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamQBPass_RateRoll'] = rolltemp.iloc[:,-1]
            roll['teamRB1_Rush_Yds'] = subteam.apply(lambda row: teamRB1_Rush_Yds(row), axis = 1)
            rolltemp['temp'] = roll['teamRB1_Rush_Yds']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamRB1_Rush_YdsRoll'] = rolltemp.iloc[:,-1]
            roll['teamRB1_Lng'] = subteam.apply(lambda row: teamRB1_Lng(row), axis = 1)
            rolltemp['temp'] = roll['teamRB1_Lng']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamRB1_LngRoll'] = rolltemp.iloc[:,-1]
            roll['teamWR1_Tgt'] = subteam.apply(lambda row: teamWR1_Tgt(row), axis = 1)
            rolltemp['temp'] = roll['teamWR1_Tgt']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['TteamWR1_TgtRoll'] = rolltemp.iloc[:,-1]
            roll['teamWR1_Rec'] = subteam.apply(lambda row: teamWR1_Rec(row), axis = 1)
            rolltemp['temp'] = roll['teamWR1_Rec']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamWR1_RecRoll'] = rolltemp.iloc[:,-1]
            roll['teamWR1_Yrds'] = subteam.apply(lambda row: teamWR1_Yrds(row), axis = 1)
            rolltemp['temp'] = roll['teamWR1_Yrds']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamWR1_YrdsRoll'] = rolltemp.iloc[:,-1]
            roll['teamWR1_TDs'] = subteam.apply(lambda row: teamWR1_TDs(row), axis = 1)
            rolltemp['temp'] = roll['teamWR1_TDs']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamWR1_TDsRoll'] = rolltemp.iloc[:,-1]
            roll['teamWR2_Tgt'] = subteam.apply(lambda row: teamWR2_Tgt(row), axis = 1)
            rolltemp['temp'] = roll['teamWR2_Tgt']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamWR2_TgtRoll'] = rolltemp.iloc[:,-1]
            roll['teamWR2_Yrds'] = subteam.apply(lambda row: teamWR2_Yrds(row), axis = 1)
            rolltemp['temp'] = roll['teamWR2_Yrds']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamWR2_YrdsRoll'] = rolltemp.iloc[:,-1]
            roll['teamWR2_TDs'] = subteam.apply(lambda row: teamWR2_TDs(row), axis = 1)
            rolltemp['temp'] = roll['teamWR2_TDs']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamWR2_TDsRoll'] = rolltemp.iloc[:,-1]
            roll['team1stDown'] = subteam.apply(lambda row: team1stDown(row), axis = 1)
            rolltemp['temp'] = roll['team1stDown']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['team1stDownRoll'] = rolltemp.iloc[:,-1]
            roll['team3rdDown'] = subteam.apply(lambda row: team3rdDown(row), axis = 1)
            rolltemp['temp'] = roll['team3rdDown']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['team3rdDownRoll'] = rolltemp.iloc[:,-1]
            roll['teamTimePoss'] = subteam.apply(lambda row: teamTimePoss(row), axis = 1)
            rolltemp['temp'] = roll['teamTimePoss']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamTimePossRoll'] = rolltemp.iloc[:,-1]
            roll['teamPenalties'] = subteam.apply(lambda row: teamPenalties(row), axis = 1)
            rolltemp['temp'] = roll['teamPenalties']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamPenaltiesRoll'] = rolltemp.iloc[:,-1]
            roll['teamPenaltiesYrd'] = subteam.apply(lambda row: teamPenaltiesYrd(row), axis = 1)
            rolltemp['temp'] = roll['teamPenaltiesYrd']
            rolltemp['temp2'] = rolltemp.expanding(min_periods=4).mean()
            subteam['teamPenaltiesYrdRoll'] = rolltemp.iloc[:,-1]
            yeardf = yeardf.append(subteam, ignore_index=True, sort = False)
        except ValueError:
            pass
    ROLLING_MEAN = ROLLING_MEAN.append(yeardf, ignore_index=True, sort = False)
```
